```python
df = pd.read_csv('./data_analyst_job_posting.csv')
df.head()
```

```python
df.info()
```

```python
df = df[['title', 'description', 'description_tokens']].copy()
df.head()
```

```python
df['title'].unique()
```

```python
target_roles = ['data analyst', 'senior data analyst']

df = df[
    df['title']
    .str.strip()
    .str.lower()
    .isin(target_roles)
].copy()
df
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
```

```python
# Role Normalization

def map_role(title):
    title = title.lower()
    
    if 'data analyst' in title:
        return 'data analyst'
    elif 'frontend' in title or 'front end' in title or 'react' in title or 'front-end' in title:
        return 'frontend'
    elif "backend" in title or 'back end' in title or "node" in title or 'back-end' in title:
        return "backend"
    else:
        return None
df['role'] = df['title'].apply(map_role)
df = df[df['role'].notnull()].copy()
df
```

```python
print(type(df['description_tokens'].iloc[0]))
print(df['description_tokens'].iloc[0])
```

```python
import re
import ast

def casefolding(text):
    return text.lower()

def cleaning(text):
    text = re.sub(r"http\S+", "", text)
    
    text = text.replace('\n', ' ')
    text = text.replace('*','')
    
    text = text.strip()
    
    text = re.sub(r'\s+', ' ', text)
    
    return text

df['description_tokens'] = df['description_tokens'].apply(ast.literal_eval)
df['description_tokens'] = df['description_tokens'].apply(lambda a:', '.join(item for item in a))

for col in df.columns:
    df.loc[:, f"clean_{col}"] = df[col].apply(casefolding)
    df.loc[:, f"clean_{col}"] = df[f"clean_{col}"].apply(cleaning)

df
```
