# Preprocessing Internshala Jobs Dataset Folder 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
id = '1aOvtU7cv-7lVE4GPIe6fEmF2SjJaCDoO'
url = f'https://drive.google.com/uc?id={id}'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(columns=['Company_Name', 'Stipend', 'Links'], inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
```

```python
# Role Normalization

def map_role(title):
    title = title.lower()
    
    if 'data analyst' in title:
        return 'data analyst'
    elif 'frontend' in title or 'front end' in title or 'react' in title or 'front-end' in title:
        return 'frontend'
    elif "backend" in title or 'back end' in title or "node" in title or 'back-end' in title:
        return "backend"
    else:
        return None
df['role'] = df['JobTitles'].apply(map_role)
df = df[df['role'].notnull()].copy()
df
```

```python
df.info()
```

```python
import re

def casefolding(text):
    return text.lower()

def cleaning(text):
    text = re.sub(r"http\S+", "", text)
    
    text = text.replace('\n', ' ')
    text = text.replace('*','')
    
    text = text.strip()
    
    text = re.sub(r'\s+', ' ', text)
    
    return text

for col in df.columns:
    df.loc[:, f"clean_{col}"] = df[col].apply(casefolding)
    df.loc[:, f"clean_{col}"] = df[f"clean_{col}"].apply(cleaning)
df
```

```python
df['role'].value_counts()
```

```python
df_final = df[['role', 'clean_Skills', 'clean_Description']].copy()
df_final
```

```python
df_final.info()
```

```python
df_final.rename(columns={'clean_Skills': 'skills', 'clean_Description': 'description'}, inplace=True)
df_final
```

```python
def preprocessing_lanjutan(text):
    text = text.replace('/', ' ')
    text = text.replace('&', ' ')
    text = re.sub(r"[()]", " ", text)
    
    return text
    
df_final['skills'] = df_final['skills'].apply(preprocessing_lanjutan).copy()
df_final
```

```python
df_final['description'] = df_final['description'].astype(str).str.strip()
df_final
```

```python
df_final.to_csv('data7.csv', index=False)
```
