# Preprocessing Resume Dataset Folder  (Pertimbangan)



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
id = '1d2WUxxfgbZ9K_yjxc5PNBZhlKsCaFSEL'
url = f'https://drive.google.com/uc?id={id}'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
```

```python
# Role Normalization

def map_role(title):
    title = title.lower()
    
    if 'data analyst' in title:
        return 'data analyst'
    elif 'frontend' in title or 'front end' in title or 'react' in title or 'front-end' in title:
        return 'frontend'
    elif "backend" in title or 'back end' in title or "node" in title or 'back-end' in title:
        return "backend"
    else:
        return None
df['role'] = df['Category'].apply(map_role)
df = df[df['role'].notnull()].copy()
df
```

```python
df.info()
```

```python
import re

def casefolding(text):
    return text.lower()

def cleaning(text):
    text = re.sub(r"http\S+", "", text)
    
    text = text.replace('\n', ' ')
    text = text.replace('*','')
    
    text = text.strip()
    
    text = re.sub(r'\s+', ' ', text)
    
    return text

for col in df.columns:
    df.loc[:, f"clean_{col}"] = df[col].apply(casefolding)
    df.loc[:, f"clean_{col}"] = df[f"clean_{col}"].apply(cleaning)
df
```
