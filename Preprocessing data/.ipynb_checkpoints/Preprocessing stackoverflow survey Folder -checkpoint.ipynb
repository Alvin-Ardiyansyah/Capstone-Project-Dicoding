# Preprocessing stackoverflow survey Folder 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
id = '1ElgvODvXhWuWPe4hnr90J3u0SLBlgl0I'
url = f'https://drive.google.com/uc?id={id}'
df = pd.read_csv(url)
df.head()
```

```python
skill_cols = [
    'LanguageHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'MiscTechHaveWorkedWith',
    'ToolsTechHaveWorkedWith'
]

df = df.copy()

df['skills'] = df[skill_cols].fillna('').agg('; '.join, axis=1)
df.head()
```

```python
df = df[['DevType', 'skills']].copy()
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
```

```python
# Role Normalization

def map_role(title):
    title = title.lower()
    
    if 'data analyst' in title:
        return 'data analyst'
    elif 'frontend' in title or 'front end' in title or 'react' in title or 'front-end' in title:
        return 'frontend'
    elif "backend" in title or 'back end' in title or "node" in title or 'back-end' in title:
        return "backend"
    else:
        return None
df['role'] = df['DevType'].apply(map_role)
df = df[df['role'].notnull()].copy()
df
```

```python
import re

def casefolding(text):
    return text.lower()

def cleaning(text):
    text = re.sub(r"http\S+", "", text)
    
    text = text.replace('\n', ' ')
    text = text.replace('*','')
    text = text.replace(';',', ')
    text = re.sub(r',\s*,+', ',', text)
    text = text.strip()
    
    text = re.sub(r'\s+', ' ', text)
    
    return text

for col in df.columns:
    df.loc[:, f"clean_{col}"] = df[col].apply(casefolding)
    df.loc[:, f"clean_{col}"] = df[f"clean_{col}"].apply(cleaning)
df
```
