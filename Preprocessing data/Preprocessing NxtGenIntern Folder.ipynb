# Preprocessing NxtGenIntern Folder

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
id = '1CbA-_VFqn4NsYQShEWBJvk8AS226eTWt'
url = f'https://drive.google.com/uc?id={id}'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Certifications'], inplace=True)
df.head()
```

```python
# Proses data null + duplicate

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.info()
```

```python
# Role Normalization

def map_role(title):
    title = title.lower()
    
    if 'data analyst' in title:
        return 'data analyst'
    elif 'frontend' in title or 'front end' in title or 'react' in title or 'front-end' in title:
        return 'frontend'
    elif "backend" in title or 'back end' in title or "node" in title or 'back-end' in title:
        return "backend"
    else:
        return None
df['role'] = df['Job Title'].apply(map_role)
df = df[df['role'].notnull()]
df
```
